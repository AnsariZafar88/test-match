```python

plt.figure(figsize=(8, 5))
sns.lineplot(x='Runs', y='Player', data=df)
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Country', data=df)
plt.yscale('log')
plt.show()
```

```python
top_players = df.nlargest(10, 'Runs')

plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Player', data=top_players, palette='viridis')
plt.title('Top 10 Players by Total Runs')
plt.xlabel('Total Runs')
plt.ylabel('Player')
plt.show()
```

```python
top_players = df.nlargest(10, 'Runs')
plt.figure(figsize=(10, 6))
sns.histplot(df['Runs'], kde=True, bins=20, color='blue')
plt.title('Distribution of Total Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Strike_Rate', y='Ave', data=df, hue='Runs', size='Runs', sizes=(20, 200), palette='cool')
plt.title('Strike Rate vs Batting Average')
plt.xlabel('Strike Rate')
plt.ylabel('Batting Average')
plt.legend(title='Total Runs', loc='upper left')
plt.show()
```

```python
century_players = df.nlargest(20, '100')

plt.figure(figsize=(12, 6))
sns.barplot(x='100', y='Player', data=century_players, palette='plasma')
plt.title('Top 20 Players by Number of Centuries')
plt.xlabel('Number of Centuries')
plt.ylabel('Player')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='6s', y='Player',  data=top_players, palette='viridis')
plt.title('Top 10 Players by Total 6s')
plt.xlabel('Total 6s')
plt.ylabel('Player')
plt.show()
```
